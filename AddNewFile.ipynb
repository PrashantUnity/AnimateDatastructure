```
using System;
using System.Collections.Generic;
using System.Text;
using System.Linq;
using System.IO;
```

```
var fileName = "equal-sum-grid-partition-ii";
var folderName = "Matrix";
```

```
var dir = Directory.GetCurrentDirectory();
```

```
var newFileAndFolderPath = Path.Combine(dir , folderName);
newFileAndFolderPath
```

```
public class ProblemData
{
    public string ProblemUri { get; set; }
    public string ProblemName { get; set; }
    public string ProblemDescription { get; set; }
    public string ProblemSolution { get; set; }
    public string Difficulty { get; set; }
    public List<string> Topics { get; set; }
    public int ProblemNumber { get; set; }
}
```

```
Directory.CreateDirectory(newFileAndFolderPath);

var jsonPath = Path.Combine(newFileAndFolderPath, fileName + ".json");
var htmlPath = Path.Combine(newFileAndFolderPath, fileName + ".html");

if (!File.Exists(jsonPath))
{
    var data = new ProblemData
    {
        ProblemUri = $"https://leetcode.com/problems/{fileName}",
        ProblemName = "Equal Sum Grid Partition II",
        ProblemDescription = """
You are given an m x n matrix grid of positive integers. Your task is to determine if it is possible to make either one horizontal or one vertical cut on the grid such that:

    Each of the two resulting sections formed by the cut is non-empty.
    The sum of elements in both sections is equal, or can be made equal by discounting at most one single cell in total (from either section).
    If a cell is discounted, the rest of the section must remain connected.

Return true if such a partition exists; otherwise, return false.

Note: A section is connected if every cell in it can be reached from any other cell by moving up, down, left, or right through other cells in the section.
""",
        ProblemSolution = "Prefix Sum",
        Difficulty = "Medium",
        Topics = """

Principal
Array
Hash Table
Matrix
Enumeration
Prefix Sum
Weekly Contest 449

""".Split([' ','\n']).Select(x=>x.Trim()).Where(x=>x.Length>0).ToList(),
        ProblemNumber = 3548
    };

    var json = System.Text.Json.JsonSerializer.Serialize(data, new System.Text.Json.JsonSerializerOptions
    {
        WriteIndented = true
    });

    File.WriteAllText(jsonPath, json);
}

if (!File.Exists(htmlPath))
{
    File.WriteAllText(htmlPath, "");
}
```
